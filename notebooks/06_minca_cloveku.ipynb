Podúloha: Aká je šanca, že počet hláv z $n$ hodov mincou bude presne $k$?



#### Riešenie podúlohy

Pravdepodobnosť spočítame, ako počet vyhovujúcich možností vydelených počtom všetkých možností.

Počet všetkých možností, ako môže dopadnúť $n$ hodov mincou, bude $2^n$. Každá minca môže dopadnúť dvoma rôznymi spôsobmi a hádžeme dokopy $n$-krát.

Počet vyhovujúcich možností je počet možností, keď hodíme $k$-krát hlavu z $n$-hodov. Počet týchto možností je rovnaký, ako odpoveď na otázku, koľko spôsobmi môžme vybrať $k$-guličiek z $n$-guličiek, čoho odpoveďou je ${n} \choose{k}$.

Odpoveďou na podúlohu je teda 

$$
\frac{n \choose k}{2^n}
$$

#### Pokračovanie

Keď máme spočítané pravdepodobnosti na počty hláv, môžme si ich vyobraziť do grafu.

````
import math

N = 100
N_RESPONDENTS = 500
# histogram spravený rovnako ako v predchádzajúcej časti
trace_histogram = go.Histogram(
    x=data.apply(lambda col: (col == 1).sum()),
    histnorm='probability',
    name='histogram') 
# spočítame vzorec pre každú z hodnôt
x_true_probs = list(range(N + 1))
y_true_probs = [math.factorial(N) / math.factorial(N-k) / math.factorial(k) / (2 ** N) 
                for k in x_true_probs]
# vykreslíme ako čiaru
trace_true_probs = go.Scatter(
    x=x_true_probs,
    y=y_true_probs,
    mode='lines',
    name='true probability'
)
fig = go.Figure(data=[trace_histogram, trace_true_probs])
fig.update_layout(
    title="Rozdelenie počtu vrchov v rámci respondentov",
    xaxis_title="Počet vrchov",
    yaxis_title="Pravdepodobnosť / zmeraná pravdepodobnosť",
)
fig.show()
````

#### Riešenie podúlohy

Ak ste sa zamysleli nad týmto problémom, pravdepodobne ste dospeli k záveru, že to vôbec nie je také jednoduché, ako sa na prvý pohľad zdá. To je správne pozorovanie. Vyriešenie tohto príkladu čisto s použitím matematiky je celkom náročné. 

Kľúčom k riešeniu tejto úlohy je matematická indukcia. Predstavte si, že máte riešenie pre $n$. Viete na základe neho nájsť riešenie pre $n+1$?

#### Riešenie podúlohy II.

Úloha je aj s použitím matematickej indukcie stále celkom náročná, ale už sa dá doriešiť. Aby sme vedeli rozumne spraviť krok z $n$ na $n+1$, potrebujeme si zapamätať výsledky pre všetky $k$ (počty dvoch hláv idúcich po sebe) a navyše aj ich rozdelenie podľa toho, *či bol posledný hod hlava alebo znak*. Prečo? Skúsme si spraviť prvých pár krokov a rýchlo to uvidíme.

* `n=1`, máme dve postupnosti s 0 dvojicami hlava-hlava. Jedna z nich končí na znak a jedna z nich končí na hlavu. Tento stav si zapíšeme ako `Z0 = 1, H0 = 1` - jedna postupnosť s 0 dvojicami `HH` končiaca na znak a jedna postupnosť s 0 dvojicami `HH` končiaca na hlavu.
* `n=2`, zoberieme si stav z `n=1`. Pozrieme sa na postupnosti v košiku `Z0`:
  * Každá z týchto postupností si môže na svoj koniec pridať buď hlavu alebo znak. Pozrieme sa na to individuálne:
    * Ak si `Z0` pridá na koniec hlavu, bude z nej `H0`. Nepridala sa jej žiadna dvojica `HH`, ale už má na konci hlavu. Z postupnosti tvaru `...xxxZ` sa stala `...xxxZH`, čo sa nám zaradí do kategórie `...xxxH`. 
    * Po pridaní znaku na koniec `Z0` dostaneme zase iba `Z0`.
  * Rovnako si rozoberieme kategóriu `H0`:
    * Ak si k nej pridáme hlavu, dostaneme sa do `H1`. Postupnosť `...xxxH` sme si rozšírili na `...xxxHH`, čím sa nám vytvorila jedna nová dvojica `HH` a stále máme na konci hlavu. 
    * Pridanie znaku na koniec `H0` vytvorí `Z0`
  * Ak si to celé zhrnieme, tak na dosiahnutie `n=2` spravíme operácie 
  ```
  Z0 <- Z0 + H0,
  H0 <- Z0,
  H1 <- H0.
  ```
  Ľavé strany týchto operácií zodpovedajú novému počtu postupností v košíkoch, kým pravé strany týchto operácií nám dávajú návod, ako ľavé strany spočítať. Použitím tohto návodu dostaneme stav `Z0 = 2, H0 = 1, H1 = 1`. Tento stav zodpovedá pravdepodobnosti $\tfrac{3}{4}$, že bude 0 dvojíc `HH` a pravdepodobnosti `\tfrac{1}{4},` že bude 1 dvojica `HH`.
* `n=3`, spravíme si ďalší krok. Rozoberte si všetkých 8 možností sami a mali by ste sa dostať k operáciám 
```
Z0 <- Z0 + H0,
H0 <- Z0,
H1 <- H0,
Z1 <- H1,
H2 <- H1
```
z čoho dostaneme stav `Z0 = 3, H0 = 2, H1 = 1, Z1 = 1, H2 = 1`. Tento stav zodpovedá pravdepodobnostnému rozloženiu $\tfrac{5}{8}$ pre 0 dvojíc, $\tfrac{2}{8}$ pre 1 dvojicu a $\tfrac{1}{8}$ pre 3 dvojice.
* `n=4`, spravíme si spoločne ešte jeden posledný krok. Skontrolujte si, že vám operácie a pravdepodobnosti vyšli takto:
```
Z0 <- Z0 + H0 = 5 
H0 <- Z0      = 3,
H1 <- H0 + Z1 = 3,
Z1 <- H1 + Z1 = 2,
H2 <- H1      = 1,
Z2 <- H2      = 1
H3 <- H2      = 1
====================
0 -> 8 / 16,
1 -> 5 / 16,
2 -> 2 / 16,
3 -> 1 / 16.
```

Tento postup nám síce nedáva jasný matematický výsledok, ale dáva nám recept, ako si vypočítať rozloženie pravdepodobností pre `n+1` na základe predchádzajúceho. Ak sa lepšie zapozeráme, uvidíme tam nasledovný vzorec:

$$
H_{k}(n+1) = H_{k-1}(n) + Z_{k}(n) \\
Z_{k}(n+1) = H_{k}(n) + Z_{k}(n),
$$
kde $n$ je dĺžka sekvencie a $k$ je počet dvojíc `HH`. Aj keď nevieme vzorcom vyjadriť, ako bude vyzerať ľubovoľný člen tejto postupnosti, môžme si na to napísať jednoduchý program, ktorý to zráta za nás.

````
def distribution_of_HH(N):
    if N <= 0:
        return {}
    if N == 1: # Pre malé N vrátime presný výsledok
        return {0: {'H': 1, 'Z': 1}} # stav v dĺžke 1
    
    # predchádzajúce výsledky, s ktorými urobíme indukčný krok
    # volanie samotnej funkcie z tela funkcie sa nazýva rekurzia
    # musíme si dať pozor, aby nám nevznikol nekonečný cyklus, o čo sme sa v našom prípade postarali
    previous_results = distribution_of_HH(N-1) 
    
    # pomocná funkcia na čítanie predchádzajúcich výsledkov
    def get_result(k, coin): # priklad štruktúry results je vyššie v časti N == 1.
        if k in previous_results:
            if coin in previous_results[k]:
                return previous_results[k][coin]
        return 0

    new_results = {} # sem budeme ukladať priebežné výsledky
    max_k = max(previous_results.keys()) + 1 # uvažujeme o jeden väčší počet dvojíc ako minulé max
    for k in range(max_k + 1): # po maximum vrátane
        resultH = get_result(k-1, 'H') + get_result(k, 'Z')
        resultZ = get_result(k, 'H') + get_result(k, 'Z')
        if resultH > 0 or resultZ > 0:
            new_results[k] = {'H': resultH, 'Z': resultZ}
    return new_results
````

````
def probabilities_of_HH(N):
    dist = distribution_of_HH(N)
    normalizer = 2 ** N
    return [(dist[key]['H'] + dist[key]['Z']) / normalizer for key in range(len(dist))]

probabilities_of_HH(4)
````

#### Overovanie výsledkov



Na štatistike je krásne, že aj keby sme nevedeli presne spočítať výsledok predchádzajúcej úlohy, môžme si ho veľmi jednoducho nasimulovať. Konkrétne si môžme začať vytvárať náhodné postupnosti dĺžky `n`, počítať na nich našu štatistiku a vykresliť si rozdelenie, ktoré sme z takéhoto experimentu dostali. Ak spravíme dostatočne veľa pokusov, mali by sme dostať riešenie veľmi podobné tomu, čo sme si spočítali. Poďme to hneď vyskúšať:

````
def count_HH(seq): ## spočítame počet dvojíc HH, teda v našich dátach 1,1
    return sum(1 if seq[i] == 1 and seq[i+1] == 1 else 0 for i in range(len(seq)-1))

def random_seq(n): ## vytvoríme náhodnú sekvenciu
    return np.random.randint(0, 2, n)

def random_HH_stat(n, samples=1000): ## spočítame našu štatistiku na náhodných dátach
    return [count_HH(random_seq(n)) for i in range(samples)]
````

````
N = 100

trace_histogram = go.Histogram(
    x=random_HH_stat(N, 10000),
    histnorm='probability',
    name='random histogram') 

probs = probabilities_of_HH(N)
trace_true_probs = go.Scatter(
    x=list(range(len(probs))),
    y=probs,
    mode='lines',
    name='true probability'
)
fig = go.Figure(data=[trace_histogram, trace_true_probs])
fig.update_layout(
    title="Rozdelenie počtu dvojíc HH",
    xaxis_title="Počet dvojíc HH",
    yaxis_title="Pravdepodobnosť",
)
fig.show()
````

Do budúcnosti teda nemusíme prechádzať ťažkým matematickým cvičením rátania zložitej pravdepodobnosti, ale za pár riadkov programu si môžme celkom presne simulovať výsledok. Takáto simulácia sa odborne nazýva *Monte-carlo simulácia*, ak by ste chceli zamachrovať pred kamarátmi.

Prístup, akým sme spočítali tieto skutočné pravdepodobnosti, sa nazýva *dynamické programovanie*. Je to oblasť, kde sa pekne mixuje matematika a programovanie. Museli sme sa chvíľu matematicky zamyslieť, aké vzťahy platia v našom matematickom probléme, aby sme prišli k rekurzívnemu / indukčnému pravidlu. Toto pravidlo sme následne priamočiaro naprogramovali a dostali sme sa rýchlo k výsledku, ktorý by sme ručne počítali veľmi dlho.

Aj dynamické programovanie aj monte-carlo simulácie sú oba skvelým príkladom toho, že programovanie je skvelým a potrebným nástrojom matematikov.

### Pokročilá štatistika


Radi by sme zopakovali náš postup z minulej časti pre viacero rôznych štatistík. Aby sme to ale nemuseli robiť celé ručne, napíšeme si kus kódu, ktorý nám to výrazne zjednoduší.

````
import plotly.figure_factory as ff
import plotly.graph_objects as go

class CoinData:
    def __init__(self):
        self.data = get_data()
    
    def compute_statistic_on_data(self, statistic):
        return self.data.apply(statistic)

    def compute_statistic_on_random(self, statistic, N=10000):
        return [statistic(np.random.randint(0, 2, len(self.data))) for i in range(N)]

    def compare_statistic(self, statistic, prob, line_height=0.1, N=10000):
        data_stats = self.compute_statistic_on_data(statistic)
        random_stats = self.compute_statistic_on_random(statistic, N)
        hist_data = go.Histogram(x=data_stats, name='data', histnorm='probability')
        hist_random = go.Histogram(x=random_stats, name='random', histnorm='probability')
        fig = go.Figure([hist_data, hist_random])

        # significances
        low, high = CoinData.get_significances(random_stats, prob)
        fig.add_traces(CoinData.traces_vertical_significances(low, high, prob, line_height))
        fig.update_layout(
            title="Rozdelenie štatistiky",
            xaxis_title="Štatistika",
            yaxis_title="Pravdepodobnosť",
        )
        data_outside = data_stats[np.logical_or(data_stats < low, data_stats > high)]
        print(f'Sekvencie mimo (1 - 2 x {prob}) intervalu:\n{data_outside}')
        return fig

    def get_outliers(self, statistic, prob):
        random_stats = self.compute_statistic_on_random(statistic, N)
        low, high = CoinData.get_significances(random_stats, prob)
        data_stats = self.compute_statistic_on_data(statistic)
        return data_stats[np.logical_or(data_stats < low, data_stats > high)].index

    @staticmethod
    def get_significances(data, prob):
        return np.quantile(data, prob), np.quantile(data, 1-prob)

    @staticmethod
    def traces_vertical_significances(low, high, prob, line_height):
        return [go.Scatter(x=[low, low], y=[0, line_height], mode='lines', name=f'low {prob}'),
                go.Scatter(x=[high, high], y=[0, line_height], mode='lines', name=f'high {prob}')]
````

S týmto pomocným kódom sa nám celé snaženie v minulej sekcii zredukuje na tento kus kódu.

````
cd = CoinData()
cd.compare_statistic(count_HH, 0.001)
````

Skúsme si teraz spraviť štatistiku počtu sekvencií `HZ`:

````
def count_HZ(col):
    return sum(1 if col[i] == 1 and col[i+1] == 0 else 0 for i in range(len(col) - 1))

cd.compare_statistic(count_HZ, 0.001)
````

Konečne sa nám podarilo identifikovať zaujímavý výsledok. Kým počet dvojíc `HH` ľudia zvládali spĺňať celkom dobre, s dvojicou `HZ` im to išlo horšie. Z grafu vieme vyčítať, že ľudia omnoho častejšie striedali hlavu na znak, ako by sa na náhodu patrilo.

Pre úplnosť sa pozrime ešte na striedanie opačným smerom.

````
def count_ZH(col):
    return sum(1 if col[i] == 0 and col[i+1] == 1 else 0 for i in range(len(col) - 1))

cd.compare_statistic(count_HZ, 0.001)
````

Výsledky oboch týchto štatistík sú veľmi podobné. Je asi prirodzené predpokladať, že na časté striedanie hlavy a znaku musí človek často striedať aj znak a hlavu. Skúsme zobrať náš posledný výsledok a spustiť na ňom funkciu `f1_score`, aby sme vedeli, ako na tom sme.
